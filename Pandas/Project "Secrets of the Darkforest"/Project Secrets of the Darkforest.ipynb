```python
df = pd.read_csv('https://code.s3.yandex.net/datasets/new_games.csv')
df.columns = df.columns.str.strip()
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')

# Original Number of Rows
original_rows = len(df)

# Data Cleaning
num_after_missingcleaned = 16679 
deleted_missing = original_rows - num_after_missingcleaned 
deleted_missing_percent = (deleted_missing / original_rows) * 100
num_after_duplicatecleaned = 16444 
deleted_duplicates = num_after_missingcleaned - num_after_duplicatecleaned 
deleted_duplicates_percent = (deleted_duplicates / original_rows) * 100  


# Final Number of Rows
final_rows = num_after_duplicatecleaned
total_deleted = original_rows - final_rows
total_deleted_percent = (total_deleted / original_rows) * 100
```

### <span style="color:darkviolet; font-size:18px;"> 4. Data Categorization </span> 

```python
df_actual['user_score'] = pd.to_numeric(df_actual['user_score'], errors='coerce')
# Categorization of Scores
def categorize_rating(score):
    if 8 <= score <= 10:
        return "High Score"
    elif 3 <= score < 8:
        return "Medium Score"
    elif 0 <= score < 3:
        return "Low Score"
    else:
        return "Invalid Score"  
```

```python
# Apply the Categorization Function to df_actual
df_actual['user_score_category'] = df_actual['user_score'].apply(categorize_rating)
```

```python
# Checking the Result
display("Examples of Games by Score Category user_score (2000–2013)")
display(df_actual[['name', 'year_of_release', 'user_score', 'user_score_category']].head())

display("Distribution of Games by Score Categories user_score (2000–2013)")
display(df_actual['user_score_category'].value_counts())
```

```python
# Convert critic_score to Numeric, invalid data will going to NaN
df_actual['critic_score'] = pd.to_numeric(df_actual['critic_score'], errors='coerce')
# Categorization Function for Critic Scores
def categorize_critic(score):
    if pd.isna(score):
        return "No Score"
    elif 80 <= score <= 100:
        return "High Score"
    elif 30 <= score < 80:
        return "Medium Score"
    elif 0 <= score < 30:
        return "Low Score"
    else:
        return "No Score"
```

```python
# Apply the Function
df_actual['critic_score_category'] = df_actual['critic_score'].apply(categorize_critic)
```

```python
# Checking the Result
display("Examples of Games by Score Category critic_score (2000–2013)")
display(df_actual[['name', 'year_of_release', 'critic_score', 'critic_score_category']].head())

display("Distribution of Games by Score Categories critic_score (2000–2013)")
display(df_actual['critic_score_category'].value_counts())
```

```python
# Convert to Numeric Type
df_actual['user_score'] = pd.to_numeric(df_actual['user_score'], errors='coerce')
df_actual['critic_score_category'] = df_actual['critic_score'].apply(categorize_critic)

# Categorization Function for User Scores
def categorize_user(score):
    if pd.isna(score):
        return "No Score"
    elif 8 <= score <= 10:
        return "High Score"
    elif 3 <= score < 8:
        return "Medium Score"
    elif 0 <= score < 3:
        return "Low Score"
    else:
        return "No Score"

# Categorization Function for Critic Scores
def categorize_critic(score):
    if pd.isna(score):
        return "No Score"
    elif 80 <= score <= 100:
        return "High Score"
    elif 30 <= score < 80:
        return "Medium Score"
    elif 0 <= score < 30:
        return "Low Score"
    else:
        return "No Score"

# Create New Columns for Score Categories
df_actual['user_rating_category'] = df_actual['user_score'].apply(categorize_user)
df_actual['critic_rating_category'] =df_actual['critic_score'].apply(categorize_critic)
```

```python
user_counts = df_actual.groupby('user_score_category').size()
critic_counts = df_actual.groupby('critic_score_category').size()
```

```python
display("Number of Games by User Score Categories:")
display(user_counts)
display("Number of Games by Critic Score Categories:")
display(critic_counts)
```

```python
# Group by Platform and Count Games
platform_counts = df_actual['platform'].value_counts()
```

```python
# Sort by Descending and Select Top 7 Platforms
top_7_platforms = platform_counts.head(7)
```

```python
display("Top 7 Platforms by Number of Games (2000–2013)")
display(top_7_platforms)
```
