````python
# –û–ø—Ä–µ–¥–µ–ª—è–µ–º –≥—Ä–∞–Ω–∏—Ü—ã –¥–ª—è —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
bounds = {
    'days_since_last_order': (0, 40),
    'order_frequency_month': (0, 30),
    'order_frequency_week': (0, 7),
    'avg_order_value': (0, 6000),
    'median_order_value': (0, 3000),
    'total_spent_last_month': (0, 80000),
    'total_spent_last_week': (0, 9000),
    'discount_usage_rate': (0, 1),
    'seasonal_menu_tried': (0, 1),
    'app_opens_per_week': (0, 50),
    'notifications_enabled': (0, 1),
    'review_rating_last_10': (1, 5),
    'review_rating_last_1': (1, 5),
    'app_crashes_last_month': (0, 10),
    'days_since_last_promo': (0, 100),
    'coffee_preference_change': (0, 1)
}
````

````python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ–±—Ä–µ–∑–∫–∏ –ø–æ –≥—Ä–∞–Ω–∏—Ü–∞–º 
def clip_bounds(X, bounds_dict):
    X = pd.DataFrame(X, columns=list(bounds_dict.keys()))
    for col, (min_val, max_val) in bounds_dict.items():
        if col in X.columns:
            X[col] = X[col].clip(lower=min_val, upper=max_val)
    return X
````

````python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ª–æ–≥–∞—Ä–∏—Ñ–º–∏—Ä–æ–≤–∞–Ω–∏—è –ø—Ä–∞–≤—ã—Ö —Ö–≤–æ—Å—Ç–æ–≤

def log_transform(X):
    X = pd.DataFrame(X, columns=list(bounds.keys()))
    skewed_cols = ['days_since_last_order', 'order_frequency_month', 'order_frequency_week',
                   'avg_order_value', 'total_spent_last_month', 'total_spent_last_week',
                   'days_since_last_promo']
    for col in skewed_cols:
        if col in X.columns:
            X[col] = np.log1p(X[col])
    return X
````

````python
# –ü–∞–π–ø–ª–∞–π–Ω—ã

num_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('clip_bounds', FunctionTransformer(clip_bounds, kw_args={'bounds_dict': bounds})),
    ('log_transform', FunctionTransformer(log_transform)),
    ('scaler', StandardScaler())
])

# –î–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
cat_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse=False))
])

# –û–±—ä–µ–¥–∏–Ω—è—é—â–∏–π ColumnTransformer
preprocessor = ColumnTransformer(transformers=[
    ('num', num_pipeline, numeric_features),
    ('cat', cat_pipeline, categorical_features)
])
````

````python
# –ü—Ä–æ–≤–µ—Ä–∫–∞ —Ä–∞–±–æ—Ç—ã –ø–∞–π–ø–ª–∞–π–Ω–∞

X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)

print("Processed train shape:", X_train_processed.shape)
print("Processed test shape:", X_test_processed.shape)
````

````python
class OutlierClipper(BaseEstimator, TransformerMixin):
    def __init__(self, lower_percentile=0.01, upper_percentile=0.99):
        self.lower_percentile = lower_percentile
        self.upper_percentile = upper_percentile
        
    def fit(self, X, y=None):
        X = np.array(X)
        self.lower_ = np.percentile(X, self.lower_percentile*100, axis=0)
        self.upper_ = np.percentile(X, self.upper_percentile*100, axis=0)
        return self
    
    def transform(self, X):
        X = np.array(X)
        return np.clip(X, self.lower_, self.upper_)
````

1. –ü–µ—Ä–µ—á–∏—Å–ª–∏—Ç–µ –≤—Å–µ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã, —Å –∫–æ—Ç–æ—Ä—ã–º–∏ –ø–ª–∞–Ω–∏—Ä—É–µ—Ç–µ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∏—Ä–æ–≤–∞—Ç—å.

2. –ü—Ä–æ–≤–µ–¥–∏—Ç–µ —Å–∏—Å—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–π –ø–µ—Ä–µ–±–æ—Ä –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –¥–ª—è `LogisticRegression`, –≤—ã–ø–æ–ª–Ω–∏—Ç–µ –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—é –¥–ª—è –∫–∞–∂–¥–æ–π –∫–æ–Ω—Ñ–∏–≥—É—Ä–∞—Ü–∏–∏.

3. –°–æ—Å—Ç–∞–≤—å—Ç–µ —Ç–∞–±–ª–∏—Ü—É —Å —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏.

4. –í—ã–±–µ—Ä–∏—Ç–µ –ª—É—á—à—É—é –º–æ–¥–µ–ª—å, –æ—Ä–∏–µ–Ω—Ç–∏—Ä—É—è—Å—å –Ω–∞ –∑–∞–¥–∞–Ω–Ω—É—é –º–µ—Ç—Ä–∏–∫—É –∫–∞—á–µ—Å—Ç–≤–∞.

````python
# –ì–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã

# –î–∏–∞–ø–∞–∑–æ–Ω C
C_values = [0.01, 0.1, 0.5, 1, 5, 10, 50]

results = []

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

for C in C_values:
    logreg = LogisticRegression(
        C=C,
        penalty='l2',
        class_weight='balanced',
        solver='liblinear',
        random_state=RANDOM_STATE,
        max_iter=1000
    )
    
    #scores = cross_val_score(logreg, X_train_processed_fe, y_train, cv=cv, scoring='roc_auc')
   # results.append({
       # 'C': C,
       # 'mean_ROC_AUC': scores.mean(),
       # 'std_ROC_AUC': scores.std()})
    
y_proba_cv = cross_val_predict(logreg, X_train_processed_fe, y_train, cv=cv, method='predict_proba')[:, 1]

# –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º RP_AUC
precision, recall, _ = precision_recall_curve(y_train, y_proba_cv)
rp_auc = auc(recall, precision)
    
results.append({
    'C': C,
    'RP_AUC': rp_auc
})

# –¢–∞–±–ª–∏—Ü–∞ —Å —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏
results_df = pd.DataFrame(results)
results_df = results_df.sort_values(by='RP_AUC', ascending=False)
print("RP_AUC –ø–æ C:\n", results_df)
````

–ú–æ–¥–µ–ª—å —Å—Ç–∞–±–∏–ª—å–Ω–∞ –∏ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –∏ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã.

<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ <a class="tocSkip"></h2>

–í–µ–∑–¥–µ –ø–µ—Ä–µ—Å—á–∏—Ç—ã–≤–∞–µ–º –ø–æ RP_AUC
    
–¢–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫—É –∏—Å–ø–æ–ª—å–∑—É–µ–º —Ç–æ–ª—å–∫–æ –¥–ª—è —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è —Ñ–∏–Ω–∞–ª—å–Ω–æ–π –º–æ–¥–µ–ª–∏.

````python
# 1. –°–æ–∑–¥–∞–Ω–∏–µ –Ω–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏
X_test_fe = X_test.copy()

# –ö–æ–º–±–∏–Ω–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –¥–ª—è —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏
X_test_fe['avg_order_per_freq'] = X_test_fe['avg_order_value'] / (X_test_fe['order_frequency_month'] + 1e-5)
X_test_fe['app_activity'] = X_test_fe['app_opens_per_week'] * X_test_fe['notifications_enabled']

X_test_fe['spend_change'] = X_test_fe['total_spent_last_month'] / (X_test_fe['total_spent_last_week'] + 1e-5)
X_test_fe['spend_change'] = np.clip(X_test_fe['spend_change'], 0, 10)

numeric_cols_to_transform = ['order_frequency_month', 'avg_order_value', 'total_spent_last_month']
for col in numeric_cols_to_transform:
    # –ö–≤–∞–¥—Ä–∞—Ç–Ω—ã–π –∫–æ—Ä–µ–Ω—å (—Å–≥–ª–∞–∂–∏–≤–∞–Ω–∏–µ –ø—Ä–∞–≤–æ–≥–æ —Ö–≤–æ—Å—Ç–∞)
    X_test_fe[f'{col}_sqrt'] = np.sqrt(np.clip(X_test_fe[col], a_min=0, a_max=None))
    # –õ–æ–≥–∞—Ä–∏—Ñ–º–∏—Ä–æ–≤–∞–Ω–∏–µ
    X_test_fe[f'{col}_log'] = np.log1p(np.clip(X_test_fe[col], a_min=0, a_max=None))
````

````python
# 2. –§–∏–Ω–∞–ª—å–Ω—ã–π –ø–∞–π–ø–ª–∞–π–Ω —Å –æ–±—Ä–∞–±–æ—Ç–∫–æ–π –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏ –º–æ–¥–µ–ª—å—é
final_pipeline = Pipeline([
    ('preprocessor', preprocessing_pipeline_fe),
    ('classifier', LogisticRegression(
        C=1, penalty='l2', class_weight='balanced', solver='liblinear', max_iter=1000, random_state=RANDOM_STATE
    ))
])
````

````python
# 3. –û–±—É—á–µ–Ω–∏–µ –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ —Å –Ω–æ–≤—ã–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏
final_pipeline.fit(X_train_fe, y_train)
````

````python
# 4. –û—Ü–µ–Ω–∫–∞ –º–æ–¥–µ–ª–∏ —á–µ—Ä–µ–∑ –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—é –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
y_proba_cv = cross_val_predict(final_pipeline, X_train_fe, y_train, cv=cv, method='predict_proba')[:, 1]

precision, recall, _ = precision_recall_curve(y_train, y_proba_cv)
rp_auc_cv = auc(recall, precision)
print("Final model RP_AUC (CV):", rp_auc_cv)
````

````python
# 5. –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
y_pred_proba_final = final_pipeline.predict_proba(X_test_fe)[:, 1]
y_pred_final = final_pipeline.predict(X_test_fe)

precision_test, recall_test, _ = precision_recall_curve(y_test, y_pred_proba_final)
rp_auc_test = auc(recall_test, precision_test)
print("Final Test RP_AUC:", rp_auc_test)
````

````python
# 6. –ö–æ–Ω—Ç—Ä–æ–ª—å —Ç–æ—á–Ω–æ—Å—Ç–∏/–ø–æ–ª–Ω–æ—Ç—ã –Ω–∞ —Ç–µ—Å—Ç–µ
report_final = classification_report(y_test, y_pred_final)
print("Final Test Classification Report:\n", report_final)
````
